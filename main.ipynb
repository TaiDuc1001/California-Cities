## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read csv file

```python
cities = pd.read_csv('./california_cities.csv')
```

### Show dataframe preview

```python
cities.head()
```

## Create range for legend

```python
area_range = [50, 100, 300, 500]
```

### View themes

```python
plt.style.available
```

## Main process

```python
lat, lon = cities['latd'], cities['longd']
population, area= cities['population_total'], cities['area_total_km2']

plt.style.use('dark_background')
plt.figure(figsize=(8,7))
plt.scatter(lon, lat,
            c=np.log10(population),
            cmap='inferno',
            alpha=0.5,
            s=area, linewidths=0)

    
plt.colorbar(label='Log$_{10}$(Population)')
plt.axis('equal')
plt.xlabel('Longtitude')
plt.ylabel('Latitude')
plt.title('California\'s Cities With Population and Area')
plt.clim(3,6)


for area in area_range:
    plt.scatter([], [], s=area, label=str(area)+' km$^2$', alpha=0.5)
    
plt.legend(title='Cities Area', labelspacing=1, frameon=False)
plt.savefig('California_cities_distribution.png')
plt.show();
```
